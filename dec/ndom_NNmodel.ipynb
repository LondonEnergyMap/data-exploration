```python
import pandas as pd

decpath = 'https://media.githubusercontent.com/media/LondonEnergyMap/cleandata/master/dec/dec_ldn.csv'
# epcpath = 'https://media.githubusercontent.com/media/LondonEnergyMap/cleandata/master/epc/nondomestic/epcldn.csv'

# read csv into dataframe
dfdec = pd.read_csv(decpath)
# dfepc = pd.read_csv(epcpath)
```

```python
dfdec.columns
```

```python
dfdec = dfdec[(dfdec.annual_thermalfuel != 0) & (dfdec.annual_elec != 0)]
```

```python
keepcols = ['add1',
            'add2',
            'add3',
            'pcode',
#             'bref',
            'insp_date',
            'benchmark',
            'mainheatingfuel',
            'tfa',
            'annual_thermalfuel',
            'typ_thermalfuel',
            'annual_elec',            
            'typ_elec',
            'renew_thermalfuel',
            'renew_elec',            
            'yr1_elecco2',            
            'modes',
            'localauth']

# keepcols = ['benchmark',
#             'mainheatingfuel',
#             'tfa',
#             'annual_thermalfuel',
#             'typ_thermalfuel',
#             'annual_elec',
#             'typ_elec',
#             'modes']

df = dfdec[keepcols]


df.head()

df[(df.annual_elec == 1)]

```

```python
rename_dict = {'annual_thermalfuel': 'hcons',
              'typ_thermalfuel': 'hcons_typ',
              'annual_elec': 'econs',
              'typ_elec': 'econs_typ',
              'mainheatingfuel': 'mheat'}

df.rename(columns=rename_dict, inplace=True)
```

```python
# convert all string to lower case so it doesnt impact 'unique' results

strcols = ['benchmark', 'mheat', 'modes']
df[strcols] = df[strcols].apply(lambda x: x.str.lower())
df.head()
```

```python
df.modes.unique()
```

```python
# convert modes to 4 columns: heating, mixed mode, natural ventilation, mech vent
import numpy as np

df['natvent'] = np.where(df.modes.str.contains('natural'), 1, 0)
df['mechvent'] = np.where(df.modes.str.contains('mechanical'), 1, 0)
df['mmode'] = np.where(df.modes.str.contains('mixed'), 1, 0)
df['heating'] = np.where(df.modes.str.contains('heating'), 1, 0)
df['ac'] = np.where(df.modes.str.contains('air'), 1, 0)

df.head()

```

```python
modes = df.modes.unique()

df[df.modes == modes[0]].head()
```

```python
# convert heating fuel col but remove (single) nan row first

df = df[df.mheat.notnull()]


fuel = df.mheat.unique()
# df[df.mheat.isnull()].head()
fuel
```

```python
#create dummy variable columns for fuel types and add to right of dataframe
mheat_dummies = pd.get_dummies(df.mheat)

# axis=1 indicate columns
df = pd.concat([df, mheat_dummies], axis=1)
df.head()
```

```python
#create dummy variable columns for benchmarks and add to right of dataframe

df = pd.concat([df, pd.get_dummies(df.benchmark)], axis=1)
df.head()
```

```python
df.columns
```

```python
# select columns for NN model

NNcols_hcons = ['econs',
                'econs_typ',
                'hcons',
                'hcons_typ',
                'tfa',
                'natvent', 'mechvent', 'mmode', 'heating', 'ac',
                'biomass', 'district heating', 'grid supplied electricity', 'lpg', 'natural gas',
                'oil', 'other',
                'bar, pub or licensed club', 'clinic', 'cultural activities',
                'dry sports and leisure facility', 'emergency services',
                'entertainment halls', 'fitness and health centre',
                'general accommodation', 'general office', 'general retail',
                'high street agency', 'hospital - clinical and research', 'hotel',
                'laboratory or operating theatre', 'long term residential',
                'public buildings with light usage', 'public waiting or circulation',
                'restaurant', 'schools and seasonal public buildings',
                'storage facility', 'swimming pool centre', 'university campus',
                'workshop'] #'covered car park', 


dfNN = df[NNcols_hcons]
len(dfNN.columns)
```

```python
dfNN.head()
```

```python
# d = dfNN_hcons[dfNN_hcons.econs == 0]
d = dfNN[dfNN['schools and seasonal public buildings'] == 1]
d = d[['econs',
   'econs_typ',
   'hcons',
   'hcons_typ',
   'tfa',
   'natvent',
   'mechvent',
   'mmode',
   'heating',
   'ac',
   'biomass',
   'district heating',
   'grid supplied electricity',
   'lpg',
   'natural gas',
   'oil',
   'other']]

d.dropna(subset=['econs', 'hcons'], inplace=True)
```

```python
d.econs.describe()
```

```python
d[d.econs == 1]
```

```python
# convert dataframe to numpy array for NN

dataset = d.values
```

```python
dataset
```

```python

ncols = 4
X = dataset[:, 0:ncols]
Y = dataset[:, ncols]
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)

print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(32, activation='relu', input_shape=(ncols,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train,
          batch_size=32, epochs=100,
          validation_data=(X_val, Y_val))
```
