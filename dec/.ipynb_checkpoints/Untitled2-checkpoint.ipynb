```python
import pandas as pd

decpath = 'https://media.githubusercontent.com/media/LondonEnergyMap/cleandata/master/dec/dec_ldn.csv'
# epcpath = 'https://media.githubusercontent.com/media/LondonEnergyMap/cleandata/master/epc/nondomestic/epcldn.csv'

# read csv into dataframe
dfdec = pd.read_csv(decpath)
dfdec.head()
```

```python
dfdec.columns
```

```python
dfdec['benchmark'] = dfdec.benchmark.str.lower()
```

```python
dfdec['hcons'] = dfdec.annual_thermalfuel * dfdec.tfa
dfdec['econs'] = dfdec.annual_elec * dfdec.tfa
```

```python
df = dfdec.groupby(['benchmark'], as_index=False).sum()[['benchmark', 'tfa', 'annual_thermalfuel', 'annual_elec', 'hcons', 'econs']]

df['counts'] = dfdec.groupby(['benchmark'], as_index=False).count()[['key']]
```

```python
df.head(10)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_style('darkgrid')

fig, axes = plt.subplots(1, 2, sharey=True, figsize=(12,7))

df['tfa_index'] = df.sort_values(by='tfa', ascending=False).reset_index().benchmark
sns.barplot(x='tfa', y='benchmark', data=df, ax=axes[0], order=df.tfa_index)
sns.barplot(x='counts', y='benchmark', data=df, ax=axes[1], order=df.tfa_index)

axes[0].set_ylabel('')
axes[1].set_ylabel('')

axes[0].set_xlabel('Floor area m2')
axes[1].set_xlabel('Count')

plt.subplots_adjust(left=0.07, bottom=0.15, right=0.95, top=0.9, wspace=0.1, hspace=0.2)
```

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(12,7))

df['hcons_index'] = df.sort_values(by='hcons', ascending=False).reset_index().benchmark
# df['econs_index'] = df.sort_values(by='econs', ascending=False).reset_index().benchmark

sns.barplot(x='hcons', y='benchmark', data=df, ax=axes[0], order=df.hcons_index)
sns.barplot(x='econs', y='benchmark', data=df, ax=axes[1], order=df.hcons_index)

axes[0].set_ylabel('')
axes[1].set_ylabel('')

axes[0].set_xlabel('Heating kWh')
axes[1].set_xlabel('Electricity kWh')

plt.subplots_adjust(left=0.07, bottom=0.15, right=0.95, top=0.9, wspace=0.1, hspace=0.2)
```

```python
# unify two dataframes and categories

epctypes_mapping = {'a1/a2 retail and financial/professional services': 'general retail',
                    'a3/a4/a5 restaurant and cafes/drinking establishments and hot food takeaways': 'restaurant',
                    'b1 offices and workshop businesses': 'general office',
                    'b2 to b7 general industrial and special industrial groups': 'industrial',
                    'b8 storage or distribution': 'storage facility', 
                    'c1 hotels': 'hotel',
                    'c2 residential institutions - hospitals and care homes': 'hospital - clinical and research',
                    'c2 residential institutions - residential schools': 'long term residential',
                    'c2 residential institutions - universities and colleges': 'university campus',
                    'c2a secure residential institutions': 'long term residential',
                    'c3/c4 dwelling and houses': 'general accommodation',
                    'd1 non-residential institutions - community/day centre': 'community centre',
                    'd1 non-residential institutions - crown and county courts': 'law courts',
                    'd1 non-residential institutions - education': 'schools and seasonal public buildings',
                    'd1 non-residential institutions - libraries museums and galleries': 'cultural activities',
                    'd1 non-residential institutions - primary health care building': 'fitness and health centre',
                    'd2 general assembly and leisure plus night clubs and theatres': 'entertainment halls',
                    'other': 'other', 
                    'others - car parks 24 hrs': 'carparks',
                    'others - emergency services': 'emergency services',
                    'others - miscellaneous 24hr activities': 'other',
                    'others - passenger terminals': 'public waiting or circulation',
                    'others - stand alone utility block': 'other'}


dfepc['cat'] = dfepc.types.map(epctypes_mapping)
dfepc['source'] = 'epc'

# change DECs
# only change carparks

import numpy as np
dfdec['cat'] = np.where(dfdec.benchmark.str.contains('covered'), 'carparks', dfdec.benchmark)
dfdec['source'] = 'dec'
```

```python
combined = pd.concat([dfepc, dfdec], sort=False)
combined.tfa.describe()
```

```python
temp = combined.groupby(['cat'], as_index=False).sum()[['cat', 'tfa', 'hcons', 'econs', 'cons_exb', 'cons_new']]
temp['counts'] = combined.groupby(['cat'], as_index=False).count()[['key']]


t = combined.groupby(['cat', 'source'], as_index=False).sum()[['cat', 'source', 'tfa']]

dec_tfa = t[t.source == 'dec'].drop('source', axis=1)
dec_tfa.rename(columns={'tfa': 'tfa_dec'}, inplace=True)
temp = temp.merge(dec_tfa, how='outer', left_on='cat', right_on='cat')
# epc_tfa = t[t.source == 'epc']


epc_tfa = t[t.source == 'epc'].drop('source', axis=1)
epc_tfa.rename(columns={'tfa': 'tfa_epc'}, inplace=True)
temp = temp.merge(epc_tfa, how='outer', left_on='cat', right_on='cat')

temp.head(30)

```

```python
# fig, axes = plt.subplots(1, 2, sharey=True, figsize=(12,10))

plt.figure(figsize=(10,8))
sbar = temp[['cat', 'tfa_dec', 'tfa_epc', 'tfa']]
sbar = sbar.sort_values('tfa', ascending=False).reset_index()
sbar = sbar.drop(['tfa', 'index'], axis=1)
sbar = sbar.fillna(0)

plt.bar(sbar.cat, sbar.tfa_dec)
plt.bar(sbar.cat, sbar.tfa_epc, bottom=sbar.tfa_dec)
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()

sbar
```
